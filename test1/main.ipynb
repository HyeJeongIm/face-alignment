```python
# Detectors
img_path  = './images/image_test2.jpg'
detectors = ['opencv', 'ssd', 'mtcnn']

# Load the original image using OpenCV
original_img = cv2.imread(img_path)
original_img = cv2.cvtColor(original_img, cv2.COLOR_BGR2RGB) # Convert from BGR to RGB

for detector in detectors:
    tic = time.time()
    img_list = DeepFace.extract_faces(img_path = img_path, detector_backend = detector)
    toc = time.time()

    # Assuming img_list[0]['face'] contains the detected face as an array
    detected_face = img_list[0]['face']

    # Create subplots: one for the original image and one for the detected face
    plt.figure()
    plt.subplot(1, 2, 1)
    plt.title('Original Image')
    plt.imshow(original_img)

    plt.subplot(1, 2, 2)
    plt.title(f'Detected Face with {detector}')
    plt.imshow(detected_face)

    plt.show()

    print(detector, "backend took", toc - tic, "seconds")
    print("!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!")
```
